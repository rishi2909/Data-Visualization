```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#seaborn: statistical data visulaization
sns.get_dataset_names() # these are the datasets that are provided by seaborn.
```

```python
df = sns.load_dataset('tips')
df
```

```python
#numerical data plotting >> how to plot some numerical data , by default seaborn.replot has scatter plot.
sns.relplot(x=df.total_bill, y=df.tip, data=df) #use to understand relationship b/w two variables. it is similar to scatter plot.
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df) #
```

```python
# we want to know that which one was smokers
sns.scatterplot(x=df.total_bill, y=df.tip, data=df,hue = 'smoker') # The hue parameter in sns.scatterplot() is used to color the points based on a categorical variable. This helps in visually distinguishing different groups within the data.
```

```python
#insights
# as the bill inc, the tip also inc
#highest tip was paid by a smoker
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker')
```
