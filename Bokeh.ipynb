```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()  #whichever plot you make is seen in this notebook. All the plot should render in notebook.

```

```python
from bokeh.sampledata.iris import flowers
from bokeh.plotting import figure,output_file,show
```

```python
flowers
```

```python
output_file('test1.html')
p = figure(title='test_flower')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.circle(x=flowers['petal_length'],y=flowers['petal_width'],size=8,color='blue',alpha=1)   # ✅ Fix: Defined size, color, and transparency
show(p)
```

```python
output_file('test1.html')
p = figure(title='test_flower')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.line(x=flowers['petal_length'],y=flowers['petal_width'],color='blue')   # ✅ Fix: Defined size, color, and transparency
show(p)
```

```python
x = [1,2,3,4]
y = [3,5,7,9]

output_file("line.html")
p = figure(title="line chart")
p.line(x,y)
show(p)
```

```python
x = [1,2,3,4]
y = [3,5,7,9]

output_file("scatter.html")
p = figure(title="line chart")
p.scatter(x,y,size=18,fill_color='green',legend_label="this is my data")
show(p)
```
